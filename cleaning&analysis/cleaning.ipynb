```python
df.columns = df.columns.str.strip()
df.columns = df.columns.str.replace(r'\s+', ' ', regex=True)

print(df.columns.tolist())
```

```python
categorical_cols = [
    "Industry Vertical",
    "SubVertical",
    "City Location",
    "Investors Name",
    "InvestmentType",
    "Remarks"
]

for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].fillna("Unknown")
    else:
        print(f"Column not found: {col}")
```

```python
import pandas as pd

df.columns = df.columns.str.strip()

categorical_cols = df.select_dtypes(include='object').columns

print("Categorical columns:", categorical_cols)

for col in categorical_cols:
    df[col] = df[col].fillna("Unknown")

numeric_cols = df.select_dtypes(include=['int64','float64']).columns

print("Numeric columns:", numeric_cols)

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())

print("\nMissing values after cleaning:")
print(df.isnull().sum())

df.to_csv("startup_cleaned_v1.csv", index=False)

print("\nCleaning completed successfully!")
```

```python
import pandas as pd

df = pd.read_excel("startup_cleaned_v1 (1).xlsx")

numeric_cols = [
    "Amount in USD",
    "Monthly Revenue",
    "Annual Burn Rate",
    "Customer Acquisition Cost (CAC)",
    "Customer Lifetime Value (CLV)",
    "Active Users",
    "Revenue Growth Rate (%)",
    "Funding Utilization (%)",
    "Profit Margin (%)",
    "Valuation at Funding",
    "Operating Expenses"
]

for col in numeric_cols:
    df[col] = df[col].replace("Unknown", pd.NA)

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())

df.to_excel("startup_cleaned_final.xlsx", index=False)

print("File fixed and saved as startup_cleaned_final.xlsx")
```
