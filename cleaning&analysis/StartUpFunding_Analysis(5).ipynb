# Univariate Analysis :

Univariate analysis focuses on analyzing individual variables to understand their distribution and key characteristics. 

It helps identify dominant sectors, funding patterns, and potential outliers within the dataset.

# Histogram-Plot: Funding Amount to check distribution

A histogram was plotted to examine the distribution of funding amounts.                 

This helps in understanding whether funding is normally distributed, skewed, or contains extreme values.

```python
plt.figure(figsize=(15,7))
plt.hist(df['Amount in USD'], bins=30)
plt.title("Distribution of Funding Amount")
plt.xlabel("Funding Amount (USD)")
plt.ylabel("Frequency")
plt.show()
```

# Bar-Plot : Industry Distribution Analysis

A bar plot was generated to visualize the number of startups across the top industry verticals.        

The chart displays the frequency of each industry category, allowing for a clear comparison between sectors.      

The industries were arranged in descending order based on their count to highlight the most dominant sectors

It then selects the top 10 industries with the highest number of startups using the head(10) function.

```python
top_industries = df['Industry_Vertical_Std'].value_counts().head(10)

plt.figure(figsize=(10,6))
top_industries.plot(kind='bar')

plt.title("Top 10 Industry Verticals by Number of Startups")
plt.xlabel("Industry Vertical")
plt.ylabel("Number of Startups")
plt.xticks(rotation=45)
plt.show()
```

# Analyze Startup Stage distribution :

The startup stage distribution shows how companies are spread across different funding stages such as Seed, Series A, Series B, and later rounds.

From the count plot, we can observe which stage has the highest concentration of startups,

indicating where most companies currently stand in their growth lifecycle.

```python
plt.figure(figsize=(13,7))
sns.countplot(x=df['Investment_Type_std'],
              order=df['Investment_Type_std'].value_counts().head(10).index )
plt.title("Startup Stage Distribution")
plt.xticks(rotation=45)
plt.show()
```

# Analyze Location distribution to identify startup hubs.



The location distribution analysis illustrates the proportion of startups across different cities.
The pie chart highlights the major startup hubs, 

showing which cities contribute the largest share of startups and reflecting regions with strong entrepreneurial ecosystems, 
ample funding opportunities, and well-developed infrastructure.

```python
top_cities = df['City_Location_Std'].value_counts().head(10)

plt.figure(figsize=(8,8))
plt.pie(
    top_cities,
    labels=top_cities.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.tab20.colors
)
plt.title("Startup Distribution Across Top 10 Cities")
plt.show()
```

# Box-Plot: Funding Outlier Detection

A boxplot was used to analyze the distribution of startup funding and identify extreme values.

Outliers in the plot highlight startups that received exceptionally high or low funding compared to the majority.

```python
df['Amount_log'] = np.log1p(df['Amount in USD'])  

# Boxplot using log-transformed funding
plt.figure(figsize=(10,6))
sns.boxplot(x=df['Amount_log'], color='lightcoral')
plt.title("Funding Outlier Detection (Log Scale)")
plt.xlabel("Log(Funding Amount + 1)")
plt.show()
```

# Startup Dataset - Bivariate Analysis


This notebook performs Bivariate Analysis on startup funding data to identify industry dominance, investment hubs, funding stages, and yearly trends.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Improve plot appearance
plt.style.use("ggplot")
```

## Dataset Overview

The dataset contains information about startups including:
- Industry
- City
- Investment Type
- Funding Amount
- Year

The goal is to analyze relationships between funding and different business factors.

```python


df.head()
```

## 1. Industry vs Total Funding

This analysis identifies which industries attract the highest total investment.

```python
industry_funding = df.groupby("Industry_Vertical_Std")["Amount in USD"].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
industry_funding.head(10).plot(kind="bar")
plt.title("Top 10 Industries by Total Funding")
plt.xlabel("Industry")
plt.ylabel("Total Funding (USD)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 2. Location vs Total Funding

This section identifies major startup investment hubs.

```python
city_funding = df.groupby("City_Location_Std")["Amount in USD"].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
city_funding.head(10).plot(kind="bar")
plt.title("Top 10 Cities by Total Funding")
plt.xlabel("City")
plt.ylabel("Total Funding (USD)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 3. Investment Type vs Funding Distribution

This analysis compares funding size across different investment stages.

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="Investment_Type_std", y="Amount in USD", data=df)
plt.title("Funding Distribution by Investment Type")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 4. Year-wise Funding Trend

This section examines how startup funding has changed over time.

```python
year_funding = df.groupby("Year")["Amount in USD"].sum()

plt.figure(figsize=(10,6))
year_funding.plot(kind="line", marker="o")
plt.title("Year-wise Total Funding Trend")
plt.xlabel("Year")
plt.ylabel("Total Funding (USD)")
plt.tight_layout()
plt.show()
```

## 5.Industry vs Average Funding
To determine which industries receive higher average funding per startup.
Total funding can be skewed by a few large investments. Average funding provides a clearer understanding of investor confidence per startup within each industry.

```python
import matplotlib.pyplot as plt

industry_avg = (
    df.groupby('Industry_Vertical_Std')['Amount in USD']
      .mean()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(10,6))
industry_avg.sort_values().plot(kind='barh')

plt.title("Top 10 Industries by Average Funding", fontsize=14)
plt.xlabel("Average Funding (USD)")
plt.ylabel("Industry")
plt.tight_layout()
plt.show()
```

Identify high-capital industries 

Understand capital intensity across sectors

## 6.Location vs Number of Deals

To identify major startup investment hubs based on deal count.

Number of deals reflects ecosystem activity and investor presence in a location.

```python
city_deals = (
    df['City_Location_Std']
      .value_counts()
      .head(10)
)

plt.figure(figsize=(10,6))
city_deals.sort_values().plot(kind='barh')

plt.title("Top 10 Cities by Number of Deals", fontsize=14)
plt.xlabel("Number of Deals")
plt.ylabel("City")
plt.tight_layout()
plt.show()
```

Identify top startup cities

Compare metro vs emerging ecosystems

## 7.Year vs Number of Startups Funded
To analyze how startup funding activity has changed over time
Shows market cycles, funding booms, or slowdowns.

```python
year_count = df.groupby('Year')['Startup_Name'].count()

year_count.plot(kind='line', marker='o')
plt.title("Year vs Number of Funded Startups")
plt.show()
```

Identify peak funding years

Detect investment downturns

Understand growth trajectory of the startup ecosystem

## 8.Stage vs Average Funding
To compare funding size across different startup stages.



Funding amounts typically increase as startups mature. This analysis validates that pattern.

```python
stage_avg = (
    df.groupby('Investment_Type_std')['Amount in USD']
      .mean()
      .sort_values(ascending=False)
)

plt.figure(figsize=(8,5))
stage_avg.plot(kind='bar')

plt.title("Average Funding by Investment Stage", fontsize=14)
plt.ylabel("Average Funding (USD)")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Funding progression by stage

Identify stages with highest capital concentration

## 9.Industry vs Profit Margin

To evaluate profitability differences across industries.



Some industries may generate higher margins despite receiving lower funding.

```python
industry_profit = (
    df.groupby('Industry_Vertical_Std')['Profit Margin (%)']
      .mean()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(10,6))
industry_profit.sort_values().plot(kind='barh')

plt.title("Top 10 Industries by Average Profit Margin", fontsize=14)
plt.xlabel("Average Profit Margin (%)")
plt.ylabel("Industry")
plt.tight_layout()
plt.show()
```

Identify high-margin industries

Compare capital-heavy vs profit-efficient sectors

## 10.Stage vs Revenue Growth

To analyze revenue growth trends across funding stages.



Early-stage startups may show aggressive growth, while later stages may stabilize.

```python
stage_growth = (
    df.groupby('Investment_Type_std')['Revenue Growth Rate (%)']
      .mean()
      .sort_values(ascending=False)
)

plt.figure(figsize=(8,5))
stage_growth.plot(kind='bar')

plt.title("Revenue Growth Rate by Investment Stage", fontsize=14)
plt.ylabel("Average Revenue Growth (%)")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Growth acceleration patterns

Stage-wise business expansion trends

## 10.Location vs Valuation
To compare startup valuation across different cities.



Valuation reflects investor perception and ecosystem maturity.


```python
city_val = (
    df.groupby('City_Location_Std')['Valuation at Funding']
      .mean()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(10,6))
city_val.sort_values().plot(kind='barh')

plt.title("Top 10 Cities by Average Valuation", fontsize=14)
plt.xlabel("Average Valuation (USD)")
plt.ylabel("City")
plt.tight_layout()
plt.show()
```

Identify high-valuation hubs

Understand geographic investment premium

## Conclusion

- Technology-related industries dominate startup funding.
- Major investment hubs include Bangalore and Mumbai.
- Later-stage investments receive higher funding amounts.


The startup ecosystem demonstrates strong expansion with increasing investor confidence.
